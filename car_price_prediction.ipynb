## Checking The Regression Assumptions

```python
data_cleaned2 = data_cleaned.drop(['Price'], axis=1)
```

```python
data_cleaned2.columns
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned2[['Mileage', 'EngineV', 'Year']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['Features'] = variables.columns
```

```python
vif
```

```python
#variance_inflation_factor?
```

```python
data_no_multicollinearity = data_cleaned2.drop(['Year'], axis=1)
```

## Create Dummy Variables

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first= True)
data_with_dummies
```

```python
data_preprocessed = data_with_dummies.copy()
```

## Create the _Linear Regression Model_

```python
target = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
inputs_scaled = scaler.transform(inputs)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, target, test_size=.2, random_state=42)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_predicted = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_predicted)
plt.xlabel('Targets (y_train)', size=13)
plt.ylabel('Predictions', size=13)
plt.xlim(6, 13)
plt.ylim(6, 13)
plt.show()
```

## Calculate Residuals and Model Prameters

```python
sns.kdeplot(y_train - y_predicted, fill=True)
plt.title('Residuals PDF', size=13)
```

```python
reg.score(x_train, y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summery = pd.DataFrame(inputs.columns, columns=['Features'])
reg_summery['Weights'] = reg.coef_ 
reg_summery
```

## Testing Our Model

```python
y_prediction_test = reg.predict(x_test)
```

```python
plt.scatter(y_prediction_test, y_test)
plt.xlabel('Target (y_test)', size=13)
plt.ylabel('Predictions (y_prediction_test)', size=13)
plt.xlim(6, 13)
plt.ylim(6, 13)
plt.show()
```

```python
df_performance = pd.DataFrame(np.exp(y_prediction_test), columns=['Prediction'])
```

```python
df_performance['Target'] = np.exp(y_test)
df_performance.head()
```

```python
y_test.reset_index(drop=True, inplace=True)
```

```python
df_performance['Target'] = np.exp(y_test)
df_performance.head()
```

## Calculate Percent Error

$$ Percent Error = | ( Target - Prediction ) * \frac {100 } {Target }| = \frac { Residual } { Target } * 100 $$

```python
df_performance['Residual'] = df_performance['Prediction'] - df_performance['Target']
df_performance['Percent Error'] = np.abs(df_performance['Residual']) / df_performance['Target'] * 100
```

```python
df_performance.head()
```

```python
df_performance.describe()
```

```python
pd.options.display.max_rows = 999
df_performance.sort_values(by='Percent Error', inplace=True)
df_performance
```
